## Creating a Dataframe from an Array¶

### Option 1

```python
import pandas as pd
import numpy as np
```

```python
# creating an array
data = np.array([[1,4],[2,5],[3,6]]) 
data
```

```python
# creating a dataframe
df = pd.DataFrame(data, 
                  index = ["row1",'row2','row3'],
                  columns = ['col1','col2'])
```

```python
# showing the dataframe
df
```

### Option 2

```python
data = ([[1,4],[2,5],[3,6]]) 
```

```python
# creating a dataframe
df = pd.DataFrame(data, 
                  index = ["row1",'row2','row3'],
                  columns = ['col1','col2'])
```

```python
# showing the dataframe
df
```

### Option 3

```python
# Lists used for this example
states = ['California', 'Texas', 'Florida', 'New York']
population = [300000,290000,2200000,1900000]
```

```python
# Storing Lists within a dictionary
dict_states = { 'States':states, 'Population':population}
```

```python
# Creating the dataframe
df = pd.DataFrame(dict_states, index=[f"row {i}" for i in range(len(states))]
)
```

# Assign

```python
# assign()
# when to use?
# adding multiple columns in a single line of code
# when you need to overwrite the values of an existing columns (best practice)
```

```python
# create random numbers for our new score columns
score1 = np.random.randint(1,100,1000)
score2 = np.random.randint(1,100,1000)
```

```python
# create a series using the random numbers
series1 = pd.Series(score1, index = np.arange(0,1000))
series2 = pd.Series(score2, index = np.arange(0,1000))
```

```python
# using asssign to add multiple columns
df_exams.assign(score1 = series1, score2 = series2)
# using assign creates a copy it does not overwrite the dataframe therefor you must assign it 
df_exams = df_exams.assign(score1 = series1, score2 = series2)
df_exams
```

# Insert

```python
# inserts new column at specific postion or index
df_exams.insert(1,'Test', series1)
# insert does not create a copy it directly affects the dataframe
```

```python
df_exams
```
