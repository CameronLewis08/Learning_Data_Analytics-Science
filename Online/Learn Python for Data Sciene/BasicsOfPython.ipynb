# Lists

```python
#Used to store multiple items in a single variable they are order inmuttable containers
```

```python
Countries = ['United States','India','China','Brazil']
Countries
```

```python
print(Countries[0])
print(Countries[1])
print(Countries[2])
print(Countries[3])
print("\n")
print(Countries[-1])
print(Countries[-2])
print(Countries[-3])
print(Countries[-4])
```

# Sorting A List

```python
# setting the reverse param to true will sort the list in decreasing order
numbers.sort(reverse=True)
numbers
```

# Dictionary

```python
# dictionary_name = {'key':'value'} this is a tuple and can take multiple variables with varying types
my_dict = {'key1':'value1', 'key2':'value2'}
my_data = {'Name':"Cam", 'Age':21}
my_data
```

```python
# keys() returns all keys in dictionary
my_data.keys()
```

```python
# values() returns all values in dictionary
my_data.values()
```

```python
# items() will return all the keys and values as pairs 
my_data.items()
```

```python
# dictionary[new key] = new value
my_data['height'] = 1.7
my_data
```

```python
# update({key}:value) will change a keys value
my_data.update({'height':1.9})
my_data
```

# Copying a Dictionary

```python
# copy() works the same on lists and dictionaries
new_dict = my_data.copy()
print(new_dict)
# does not copy reference only copy values so updates to the old dictionary will not affect the new dictionary
```

```python
# however when using the = operator it will
new_dict_2 = my_data
new_dict_2
```

# Removing an Element

```python
# pop(key) works similarly to lists but instead of an index you provide it a key and it will return the removed key
my_data.pop('Name')
my_data
```

```python
# del dictionary_name[key] also works the same
del my_data['Age']
my_data
```

```python
# clear() will remove all keys and values from a dictionary
my_data.clear()
my_data
```

# For Loop

for<variable> in <iterable>:
    <code>
        

```python
for country in Countries:
    print(country)
```

```python
for i, country in enumerate(Countries):
    print(i,country)
print('\n')
for country in enumerate(Countries):
    print(country)
```

```python
my_data = {'Name':"Cam", 'Age':21}
for data in my_data:
    print(data)

print('\n')

for key, value in my_data.items():
    print(key,value)
```
